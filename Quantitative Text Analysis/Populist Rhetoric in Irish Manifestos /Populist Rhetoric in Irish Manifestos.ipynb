```r
#Uploading the datafile
manifestos <- readr::read_csv(
  "path/to/your/ireland_ge_2020-24_manifestos.csv"
)
```

```r
# Exploring the datafile 
str(manifestos)
summary(manifestos)
table(manifestos$party)

# Recoding the names 
manifestos$party <- recode(manifestos$party,
  "AO"  = "Aontú",
  "FF"  = "Fianna Fáil",
  "FG"  = "Fine Gael",
  "GR"  = "Green Party",
  "II"  = "Independent Ireland",
  "LAB" = "Labour",
  "Lab" = "Labour",      
  "PBP" = "People Before Profit",
  "SD"  = "Social Democrats",
  "SF"  = "Sinn Féin"
)
table(manifestos$party, manifestos$year)
```

```r
# Adding metadata to manifestos - ideology and voteshare 

# Adding vote share by party and year
vote_data <- tibble::tibble(
  party = rep(c(
    "Sinn Féin", "Fianna Fáil", "Fine Gael", "Green Party",
    "Labour", "Social Democrats", "People Before Profit",
    "Aontú", "Independent Ireland"
  ), each = 2),
  year = rep(c(2020, 2024), times = 9),
  vote_share = c(
    24.5, 19.0,
    22.2, 21.9,
    20.9, 20.8,
    7.1,  3.0,
    4.4,  4.7,
    2.9,  4.8,
    2.6,  2.8,
    1.9,  3.9,
    NA,   3.6
  )
)

# Adding ideology by party and creating grouped ideology
ideology_data <- tibble::tibble(
  party = c(
    "Sinn Féin", "Fianna Fáil", "Fine Gael", "Green Party",
    "Labour", "Social Democrats", "People Before Profit",
    "Aontú", "Independent Ireland"
  ),
  ideology = c(
    "Left", "Centre", "Centre-right", "Green-left",
    "Centre-left", "Centre-left", "Far-left",
    "Right", "Centre-right"
  )
) |>
  mutate(
    ideology_group = case_when(
      ideology %in% c("Left", "Far-left") ~ "Left",
      ideology %in% c("Centre-left", "Green-left") ~ "Centre-left",
      ideology == "Centre" ~ "Centre",
      ideology %in% c("Centre-right", "Right") ~ "Right"
    )
  )

# Joining ideology and grouped ideology to manifestos
manifestos <- manifestos |>
  left_join(vote_data, by = c("party", "year")) |>
  left_join(ideology_data, by = "party")

ideology_data
```

```r
# Creating DFM

manifestos_dfm <- manifestos |>
  quanteda::corpus(text_field = "text") |>
  quanteda::tokens(remove_punct = TRUE, remove_numbers = TRUE) |>
  quanteda::tokens_remove(pattern = stopwords::stopwords("english")) |>
 # quanteda::tokens_wordstem() |>  
  quanteda::dfm()

# Tokens total 
sum(ntoken(manifestos_dfm))

# Tokens per year 
tokens_per_year <- dfm_group(manifestos_dfm, groups = docvars(manifestos_dfm, "year")) |>
  ntoken()
tokens_per_year

# Tokens per party 
tokens_per_party <- dfm_group(manifestos_dfm, groups = docvars(manifestos_dfm, "party")) |>
  ntoken()
tokens_per_party

# Consider dropping II, since they entered in 2023 only. 
```

```r
# Measuring Populism 

populism_dict <- dictionary(list(
  anti_elite = c("elite", "establishment", "corrupt", "mafia", "neoliberal", "privileged", "powerful", "oligarchy"),
  people_centric = c("people", "citizen", "citizens", "community", "communities", "public", "working", "ordinary", "families", "family")
))

# Applying dictionary
manifestos_pop_dfm <- dfm_lookup(manifestos_dfm, dictionary = populism_dict)

# Converting before modifying
pop_scores <- convert(manifestos_pop_dfm, to = "data.frame")

# Normalizing scores
token_totals <- ntoken(manifestos_dfm)
pop_scores$anti_elite <- pop_scores$anti_elite / token_totals * 1000
pop_scores$people_centric <- pop_scores$people_centric / token_totals * 1000

# Adding metadata
pop_scores$party <- docvars(manifestos_dfm, "party")
pop_scores$year <- docvars(manifestos_dfm, "year")

# Appending to manifestos 
manifestos <- manifestos |>
  left_join(pop_scores, by = c("party", "year"))

pop_scores
```

```r
# Measuring Sentiment

# Applying AFINN sentiment dictionary
afinn <- read_delim("path/to/your/AFINN-111.txt", delim = "\t", col_names = c("word", "value"))

# Unnesting tokens from manifesto texts 
manifestos_tidy <- manifestos |>
  unnest_tokens(word, text)

# Joining with AFINN scores and summarising sentiment by party-year
manifestos_sentiment <- manifestos_tidy |>
  inner_join(afinn, by = "word") |>
  group_by(party, year) |>
  summarise(
    sentiment_sum = sum(value),         # total sentiment score
    sentiment_avg = mean(value),        # average score per matched word
    sentiment_n = n(),                  # number of sentiment words matched
    .groups = "drop"
  )

# Merging sentiment scores back into the main manifestos dataset
manifestos <- left_join(manifestos, manifestos_sentiment, by = c("party", "year"))
```

```r
# Cosine similarity score 

# Combining all manifestos per party-year into single texts
manifesto_docs <- manifestos |>
  mutate(doc_id = paste0(party, "_", year)) |>
  group_by(doc_id) |>
  summarise(text = paste(text, collapse = " "), .groups = "drop")

# Creating DFM
manifesto_dfm <- manifesto_docs |>
  quanteda::corpus(text_field = "text", docid_field = "doc_id") |>
  quanteda::tokens(remove_punct = TRUE, remove_numbers = TRUE) |>
  quanteda::tokens_tolower() |>
  quanteda::dfm()

# Converting to matrix of term probabilities
manifesto_matrix <- quanteda::convert(manifesto_dfm, to = "matrix")
manifesto_mu <- manifesto_matrix / rowSums(manifesto_matrix)

# Getting party names 
parties <- manifestos$party[duplicated(manifestos$party)]
parties <- unique(parties)

# Calculating cosine similarity for each party between 2020 and 2024
cosine_similarity <- sapply(parties, function(p) {
  doc1 <- paste0(p, "_2020")
  doc2 <- paste0(p, "_2024")
  if (doc1 %in% rownames(manifesto_mu) && doc2 %in% rownames(manifesto_mu)) {
    sum(manifesto_mu[doc1, ] * manifesto_mu[doc2, ]) /
      (sqrt(sum(manifesto_mu[doc1, ]^2)) * sqrt(sum(manifesto_mu[doc2, ]^2)))
  } else {
    NA
  }
})

# Showing sorted similarity scores
sort(cosine_similarity, decreasing = TRUE)
```

```r
# Glove model 
manifesto_tokens <- manifesto_docs |>
  corpus(text_field = "text", docid_field = "doc_id") |>
  tokens(remove_punct = TRUE, remove_numbers = TRUE) |>
  tokens_tolower()

# Creating FCM
manifesto_fcm <- fcm(manifesto_tokens, context = "window", window = 5, tri = FALSE)

# Checking the result
print(manifesto_fcm)

# Initializing GloVe model
glove <- GlobalVectors$new(rank = 50, x_max = 10)

# Training the embeddings 
wv_main <- glove$fit_transform(
  x = manifesto_fcm,
  n_iter = 15, 
  convergence_tol = 0.01
)

# Retrieving context vectors and combining with main embeddings
wv_context <- glove$components
word_vectors <- wv_main + t(wv_context)

# Checking dimensions and example words
dim(word_vectors)
head(rownames(word_vectors), 10)

elite_vector <- word_vectors["elite", , drop = FALSE]

elite_sim <- sim2(x = word_vectors, y = elite_vector, method = "cosine", norm = "l2")
top_elite <- sort(elite_sim[,1], decreasing = TRUE)[1:15]

print(top_elite)
```

```r
# Creating the populism axis: "people" - "elite"
populism_vec <- word_vectors["people", , drop = FALSE] - word_vectors["elite", , drop = FALSE]

# Computing cosine similarity between all words and the populism axis
populism_sim <- sim2(word_vectors, populism_vec, method = "cosine", norm = "l2")

# Getting top 15 words most aligned with the populist dimension
top_populist_words <- sort(populism_sim[, 1], decreasing = TRUE)[1:15]
print(top_populist_words)
```
